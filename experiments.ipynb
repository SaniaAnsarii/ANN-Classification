```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data.head()
```

```python
label_encoder_gender = LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoded_def = pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography'])) 
```

```python
geo_encoded_def.head()
```

```python
data = pd.concat(
    [data.drop('Geography', axis=1), geo_encoded_def],
    axis=1
)
```

```python
data.head()
```

```python
onehot_encoder_geo
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
```

```python
with open('onehot_encoder_geo','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
X = data.drop('Exited',axis=1)
Y= data['Exited']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime

```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='relu')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
log_dir = 'logs/fit/'+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(
    log_dir=log_dir,
    histogram_freq=1
)
```

```python
earlystopping_callback = EarlyStopping(monitor="val_loss",patience=5,restore_best_weights=True)
```

```python
history = model.fit(
    X_train,Y_train,validation_data = (X_test,Y_test),epochs=100,callbacks=[tensorboard_callback,earlystopping_callback]
)
```

```python
model.save('model.h5')
```
